### This notebook is used to map the differences between energies for certain sequences provided by the MFold algorithm vs the Vienna Algorithm. R coef is calculated at the bottom

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
vienna = pd.read_csv("vienna_energies.csv")
mfold = pd.read_csv("mfold_energies.csv")
consolidated = pd.read_csv("consolidated_energies.csv")
```

```python
consolidated = vienna
consolidated["2"] = mfold["Energy (kcal/mol)"]
consolidated.columns = ['Sequence', 'Vienna', 'MFold'] #energies in kcal/mol
```

```python
consolidated.to_csv("consolidated_energies.csv", index=False)
```

```python
plt.scatter(consolidated['Vienna'], consolidated['MFold'])
plt.title("Vienna vs MFold Free Energy Plots")
plt.xlabel('Vienna Free Energy (kcal/mol)')
plt.ylabel('MFold Free Energy (kcal/mol)')
plt.show()
```

```python
#print(Correlation between the two)
slope, intercept, r_value, p_value, std_err = stats.linregress(consolidated['Vienna'],  consolidated['MFold'])
```

```python
print("slope: %f    r_value: %f     std_err: %f" % (slope, r_value, std_err))
```
